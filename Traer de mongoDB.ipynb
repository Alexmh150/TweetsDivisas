```python
sem1 = Jul_Week1['Text'].tolist()
len(sem1)
```

```python
sem1_text = " ".join(sem1)
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(sem1_text,"spanish")
tokens = [word.lower() for word in tokens if word.isalpha()] # Remover los signos de puntuación
for token in tokens:
    if token in ['https', 'años', 'julio', 'día', 'mira', 'ver', 'cómo', 
                 'después','hoy', 'hace', 'ser', 'saber', 'dijo', 'así',
                 'dice', 'aquí', 'tras', 'según', 'puede', 'gran', 'ahora', 
                 'debes', 'contamos', 'acuerdo', 'hacer', 'parte']:
        tokens.remove(token)
len(tokens)
```

```python
from nltk.corpus import stopwords

clean_tokens = tokens[:]
 
for token in tokens:
    if token in stopwords.words('spanish'):
        clean_tokens.remove(token)

len(clean_tokens)
```

```python
import spacy
nlp = spacy.blank('es')
```

```python
clean_tokens_sin_lem = []
separator = ' '
for token in nlp(separator.join(clean_tokens)): #Se juntan los tokens en una string, string es el formato necesario para nlp
    clean_tokens_sin_lem.append(token.lemma_)
len(clean_tokens_sin_lem)
```

```python
import nltk
freq_clean_sin_lem = nltk.FreqDist(clean_tokens_sin_lem)
for key,val in freq_clean_sin_lem.items():
    pass
    #print (str(key) + ':' + str(val))
```
